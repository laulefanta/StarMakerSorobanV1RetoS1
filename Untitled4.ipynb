```python
# Instalar Java
!apt update -y
!apt install openjdk-17-jdk -y

# Crear archivo Java
code = """
class Ciudad {
    String nombre;

    public Ciudad(String nombre) {
        this.nombre = nombre;
    }

    public String toString() {
        return nombre;
    }
}

public class RutaOptima {
    private Ciudad[] ciudades = new Ciudad[20];
    private int[] costos = new int[20];

    public void agregarRuta(Ciudad ciudad, int costo) {
        for (int i = 0; i < ciudades.length; i++) {
            if (ciudades[i] == null) {
                ciudades[i] = ciudad;
                costos[i] = costo;
                break;
            }
        }
    }

    public Ciudad[] getCiudades() {
        return ciudades;
    }

    public int[] getCostos() {
        return costos;
    }

    public static void main(String[] args) {
        RutaOptima ruta = new RutaOptima();

        ruta.agregarRuta(new Ciudad("Bogotá"), 100);
        ruta.agregarRuta(new Ciudad("Medellín"), 200);

        Ciudad[] ciudades = ruta.getCiudades();
        int[] costos = ruta.getCostos();

        for (int i = 0; i < ciudades.length; i++) {
            if (ciudades[i] != null) {
                System.out.println(ciudades[i] + " - Costo: " + costos[i]);
            }
        }
    }
}
"""

# Guardar archivo
with open("RutaOptima.java", "w") as f:
    f.write(code)

# Compilar
!javac RutaOptima.java

# Ejecutar
!java RutaOptima
```
