## Análisis Estadístico y Comparativo

### Subtask:
Calcular medidas de tendencia central, dispersión e identificar patrones u outliers en variables clave del mercado laboral de IA.


**Reasoning**:
I will calculate the descriptive statistics, identify outliers using the IQR method, and perform group comparisons by industry to detect sector-specific patterns as requested in the subtask instructions.



```python
import pandas as pd
import numpy as np

# 1. Selection of critical numerical variables
vars_interes = ['salary_before_usd', 'salary_after_usd', 'automation_risk_percent', 'salary_change_percent']

# 2. Descriptive statistics: Central tendency and dispersion
stats_desc = df[vars_interes].describe()
# Adding variance and IQR specifically
stats_desc.loc['var'] = df[vars_interes].var()
stats_desc.loc['iqr'] = stats_desc.loc['75%'] - stats_desc.loc['25%']

print("--- Medidas de Tendencia Central y Dispersión ---")
display(stats_desc)

# 3. Identifying Outliers using IQR rule
outliers_summary = {}
for col in vars_interes:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_summary[col] = len(outliers)

print("\n--- Conteo de Outliers por Variable (Regla IQR) ---")
for col, count in outliers_summary.items():
    print(f"{col}: {count} outliers detectados")

# 4. Comparison grouped by industry
sector_analysis = df.groupby('industry')[['salary_change_percent', 'automation_risk_percent']].mean().sort_values(by='salary_change_percent', ascending=False)

print("\n--- Patrones Sectoriales (Promedios por Industria) ---")
display(sector_analysis)

# 5. Patterns by job role (top 10 by automation risk)
role_analysis = df.groupby('job_role')[['automation_risk_percent', 'salary_change_percent']].mean().sort_values(by='automation_risk_percent', ascending=False)
print("\n--- Top 10 Roles con mayor Riesgo de Automatización ---")
display(role_analysis.head(10))
```

## Cálculo de Tendencia Central

### Subtask:
Calcular la media, mediana y moda para las variables numéricas clave como 'salary_after_usd', 'automation_risk_percent' y 'salary_change_percent', interpretando qué significan estos valores en el contexto del mercado laboral.


**Reasoning**:
I will calculate the mean, median, and mode for the specified numerical variables to provide a comprehensive central tendency analysis and then present them in a clear, comparative table with interpretation.



```python
import pandas as pd

# 1. Selection of key numerical columns
key_cols = ['salary_after_usd', 'automation_risk_percent', 'salary_change_percent']

# 2. Calculating Mean, Median, and Mode
means = df[key_cols].mean()
medians = df[key_cols].median()
# Mode can return multiple values, we take the first one (index 0)
modes = df[key_cols].mode().iloc[0]

# 3. Organizing results into a comparative DataFrame
central_tendency = pd.DataFrame({
    'Media (Mean)': means,
    'Mediana (Median)': medians,
    'Moda (Mode)': modes
})

print("--- Medidas de Tendencia Central ---")
display(central_tendency)

# 4. Interpretación textual
print("\n--- Interpretación de Resultados ---")
print(f"1. Riesgo de Automatización: El promedio de {central_tendency.loc['automation_risk_percent', 'Media (Mean)']:.2f}% sugiere que casi la mitad de las tareas son automatizables. La cercanía entre media y mediana indica una distribución relativamente balanceada en este riesgo.")
print(f"2. Salarios (Post-IA): El salario promedio es ${central_tendency.loc['salary_after_usd', 'Media (Mean)']:,.2f} USD. Al comparar con la mediana (${central_tendency.loc['salary_after_usd', 'Mediana (Median)']:,.2f} USD), observamos una ligera diferencia que podría indicar un leve sesgo hacia salarios más altos o bajos en ciertos sectores.")
print(f"3. Cambio Salarial: La media de {central_tendency.loc['salary_change_percent', 'Media (Mean)']:.2f}% confirma un impacto salarial macroeconómico mínimo, aunque la moda de {central_tendency.loc['salary_change_percent', 'Moda (Mode)']:.2f}% refleja el valor más frecuente observado en el dataset.")
```

## Análisis de Dispersión

### Subtask:
Calcular el rango, la varianza y la desviación estándar de las métricas salariales y de riesgo para cuantificar la variabilidad y estabilidad de los datos.


**Reasoning**:
Calculate dispersion measures (range, variance, and standard deviation) for the key metrics to quantify variability as specified in the subtask instructions.



```python
import pandas as pd

# 1. Define the list of key columns
key_cols = ['salary_after_usd', 'automation_risk_percent', 'salary_change_percent']

# 2. Calculate Standard Deviation and Variance
stds = df[key_cols].std()
variances = df[key_cols].var()

# 3. Calculate Range (Max - Min)
ranges = df[key_cols].max() - df[key_cols].min()

# 4. Consolidate results into 'dispersion_stats' DataFrame
dispersion_stats = pd.DataFrame({
    'Desviación Estándar': stds,
    'Varianza': variances,
    'Rango': ranges
})

# 5. Print the results and provide interpretation
print("--- Medidas de Dispersión ---")
display(dispersion_stats)

print("\n--- Interpretación de la Dispersión ---")
max_std_col = stds.idxmax()
print(f"La columna con mayor variabilidad es '{max_std_col}' con una desviación estándar de {stds.max():,.2f}.")
print("Esto implica una alta heterogeneidad en los salarios post-IA, sugiriendo que mientras algunos roles mantienen una compensación estable, otros experimentan fluctuaciones drásticas, lo que impacta la predictibilidad de la estabilidad laboral en el sector.")
```

## Identificación de Outliers y Relaciones

### Subtask:
Utilizar el método de Rango Intercuartílico (IQR) para detectar valores atípicos y analizar la correlación entre el riesgo de automatización y el cambio salarial.


**Reasoning**:
I will calculate the IQR for 'salary_change_percent' to identify outliers, confirm the count of 114, and then compute the Pearson correlation with 'automation_risk_percent' as instructed.



```python
import pandas as pd

# 1. Calculate Q1, Q3 and IQR for 'salary_change_percent'
Q1 = df['salary_change_percent'].quantile(0.25)
Q3 = df['salary_change_percent'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 2. Filter outliers
outliers_df = df[(df['salary_change_percent'] < lower_bound) | (df['salary_change_percent'] > upper_bound)]
print(f"Numero de outliers detectados en 'salary_change_percent': {len(outliers_df)}")

# 3. Calculate Pearson correlation
correlation = df['automation_risk_percent'].corr(df['salary_change_percent'])

print(f"\nCorrelacion de Pearson entre riesgo de automatizacion y cambio salarial: {correlation:.4f}")

# 4. Conclusion
print("\n--- Conclusion ---")
if abs(correlation) < 0.1:
    strength = "insignificante"
elif abs(correlation) < 0.3:
    strength = "debil"
else:
    strength = "moderada/fuerte"

direction = "positiva" if correlation > 0 else "negativa"
print(f"La relacion es {strength} y {direction}. Un valor cercano a 0 indica que, a nivel macro, el riesgo de automatizacion no dicta linealmente la direccion del cambio salarial en este dataset, sugiriendo que otros factores tienen mayor peso.")
```

## Crear Panel de Indicadores Clave (KPIs)

### Subtask:
Generar una visualización de métricas de alto nivel que incluya el salario promedio post-IA, el porcentaje global de riesgo de automatización y el conteo de los 114 outliers detectados.


**Reasoning**:
I will create a visualization of the high-level KPIs (mean salary, automation risk, and outlier count) using a customized Matplotlib figure to present them as executive summary cards as requested.

## Análisis de Perfiles Críticos

### Subtask:
Generar una tabla resumen visual dentro del dashboard que destaque específicamente los datos de 'Truck Driver' y 'Customer Support Rep', contrastando sus métricas con los promedios globales.


**Reasoning**:
I will calculate the global averages for automation risk and salary change, filter the data for the 'Truck Driver' and 'Customer Support Rep' roles, and create a styled comparative table highlighting their metrics against the global benchmarks.



```python
import pandas as pd

# 1. Identify global reference metrics from the main dataframe 'df'
global_avg_risk = df['automation_risk_percent'].mean()
global_avg_change = df['salary_change_percent'].mean()

# 2. Filter role_analysis for specific roles
critical_roles = ['Truck Driver', 'Customer Support Rep']
comparison_table = role_analysis.loc[role_analysis.index.isin(critical_roles),
                                     ['automation_risk_percent', 'salary_change_percent']].copy()

# 3. Add global references and calculate gaps
comparison_table['Global_Avg_Risk'] = global_avg_risk
comparison_table['Risk_Gap'] = comparison_table['automation_risk_percent'] - global_avg_risk
comparison_table['Global_Avg_Salary_Change'] = global_avg_change
comparison_table['Salary_Change_Gap'] = comparison_table['salary_change_percent'] - global_avg_change

# 4. Apply styles for highlighting critical indicators
def highlight_critical(val):
    color = 'red' if val > 0 else 'green' # Higher risk gap is bad (red), lower is good
    return f'color: {color}; font-weight: bold'

def highlight_salary_gap(val):
    color = 'red' if val < 0 else 'green' # Lower salary gap is bad (red)
    return f'color: {color}; font-weight: bold'

# Using .map() instead of .applymap() to avoid deprecation warnings
styled_table = comparison_table.style.format({
    'automation_risk_percent': '{:.2f}%',
    'salary_change_percent': '{:.2f}%',
    'Global_Avg_Risk': '{:.2f}%',
    'Risk_Gap': '{:+.2f}%',
    'Global_Avg_Salary_Change': '{:.2f}%',
    'Salary_Change_Gap': '{:+.2f}%'
}).map(highlight_critical, subset=['Risk_Gap'])\
  .map(highlight_salary_gap, subset=['Salary_Change_Gap'])\
  .set_caption("Análisis de Roles Críticos vs. Promedio Global")\
  .set_table_styles([{'selector': 'caption', 'props': [('font-size', '16px'), ('font-weight', 'bold')]}])

print("--- Tabla de Comparativa de Roles Críticos ---")
display(styled_table)
```
